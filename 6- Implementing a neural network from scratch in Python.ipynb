```python
import numpy as np
```

```python
class MLP:

  def __init__(self, num_inputs=3, num_hidden=[3, 5], num_outputs=2):
    self.num_inputs = num_inputs
    self.num_hidden = num_hidden
    self.num_outputs = num_outputs

    # Capa de entrada tendra dos neuronas
    # Habrán dos capas ocultas. La primera tendra 3 neuronas y la segunda 5 neuronas
    # Capa de salida tendra dos neuronas

    layers = [self.num_inputs] + self.num_hidden + [self.num_outputs]

    # inicializar pesos aleatorios

    self.weights = []

    # Crea matriz para cada par de capas
    for i in range(len(layers) - 1):
      w = np.random.rand(layers[i], layers[i + 1])  #Crea un arreglo de 2 dimeniones aleatorio (conexion de capa previa capa siguiente)
      self.weights.append(w)

  def forward_propagate(self, inputs):
    # inputs: vector

    activation = inputs

    for w in self.weights:
      # Calcula las entradas netas
      net_inputs = np.dot(activation, w)

      # Calcula la activacion
      activation = self._sigmoid(net_inputs)

    return activation

  def _sigmoid(self, x):
    return 1 / (1 + np.exp(-x))
```

```python
if __name__ == "__main__":

  # crear un MLP
  mlp = MLP()

  # crear algun input
  inputs = np.random.rand(mlp.num_inputs)

  # correr la propagacion forward
  output = mlp.forward_propagate(inputs)

  # mostrar resultados
  print("The network input is: {}".format(inputs))
  print("The network output is: {}".format(output))
  print("The network weights is: {}".format(mlp.weights))
```
