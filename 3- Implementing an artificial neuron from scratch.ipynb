# Primera neurona implementada en Python desde scratch


```python
import math

#Funcion sigmoide
def sigmoid(x):
  y = 1.0/(1 + math.exp(-x))
  return y

#Funcion de activacion
def activate(input,weights):
  #Realizar una entrada neta
  h = 0
  for x, w in zip(input, weights):
    h += x*w
  #Realizar la activacion
  return sigmoid(h)
```

```python
if __name__ == "__main__": #Nos aseguramos de ejecutar el script

  #Replica la estructura de la neurona que tenemos en el apunte
  input = [.5, .3, .2]
  weights = [.4, .7, .2]
  output = activate(input, weights)
  print(output)
```
